```python
import json

with open('../data/raw/nsynth-train/examples.json') as f:
    data = json.load(f)
```

```python
data
```

```python
unique_sample_rates = set()
unique_instrument_families = set()
unique_pitches = set()
unique_velocities = set()
unique_instrument_sources = set()
unique_sum_qualities = set()
unique_qualities = set()

for key, value in data.items():
    if value['instrument_family_str'] == 'bass':
        unique_sample_rates.add(value['sample_rate'])
        unique_instrument_families.add(value['instrument_family'])
        unique_pitches.add(value['pitch'])
        unique_velocities.add(value['velocity'])
        unique_instrument_sources.add(value['instrument_source'])
        unique_sum_qualities.add(sum(value['qualities']))
        for quality in value['qualities_str']:
            unique_qualities.add(quality)


# Print unique values
print("Unique Sample Rates:", unique_sample_rates)
print("Unique Instrument Families:", unique_instrument_families)
print("Unique Pitches:", unique_pitches)
print("Unique Velocities:", unique_velocities)
print("Unique Instrument Sources:", unique_instrument_sources)
print("Unique Sum Qualities:", unique_sum_qualities)
print("Unique Qualities:", unique_qualities)
```
