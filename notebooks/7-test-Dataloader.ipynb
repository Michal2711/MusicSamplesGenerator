```python
import sys
sys.path.append('../src')
from data.audio_dataset import AudioSpectrogramDataset
import matplotlib.pyplot as plt

import librosa
from torch.utils.data import DataLoader
```

```python
batch_size = 32
# base_directory = "../data/raw/Bass"
base_directory = "../data/raw/nsynth-train/audio"

dataset = AudioSpectrogramDataset(base_directory=base_directory, spectro_type='mel')
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
dataiter = iter(dataloader)
spectrograms = next(dataiter)
spectrograms.shape
```

```python
fig, axs = plt.subplots(8, 4, figsize=(150, 60))

for i, spec in enumerate(spectrograms):
    row = i // 4
    col = i % 4
    spec = spec.squeeze(0)
    img = librosa.display.specshow(spec.numpy(),
                                    x_axis='time',
                                    y_axis='log',
                                    ax=axs[row, col])
    axs[row, col].set_title(f'{i}', fontsize=20)
    fig.colorbar(img, ax=axs[row, col], format=f'%0.2f')
plt.show()
```
